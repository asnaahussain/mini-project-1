```python
import os
from sklearn.datasets import load_iris
from sklearn.preprocessing import LabelEncoder
```

```python
dataset = load_iris()
dataset_df = pd.DataFrame(data=dataset.data, columns=dataset.feature_names)
dataset_df['species'] = dataset.target_names[dataset.target]

# Apply Label Encoding
label_encoder = LabelEncoder()
dataset_df['species_encoded'] = label_encoder.fit_transform(dataset_df['species'])

# Display the first few rows of the DataFrame
print(dataset_df[['species', 'species_encoded']].head(149))
```

```python
dataset_df['species'] = dataset.target_names[dataset.target]


dataset_df['petal_area'] = dataset_df['petal length (cm)'] * dataset_df['petal width (cm)']


dataset_df['sepal_petal_length_ratio'] = dataset_df['sepal length (cm)'] / dataset_df['petal length (cm)']


print(dataset_df[['species', 'petal_area', 'sepal_petal_length_ratio']].head(149))
```

```python
from sklearn.datasets import load_iris


print("Original class distribution:")
print(dataset_df['species'].value_counts())

target_samples = dataset_df['species'].value_counts().min()
print("Target number of samples:", target_samples)

# undersampling
setosa_samples = dataset_df[dataset_df['species'] == 0]
versicolor_samples = dataset_df[dataset_df['species'] == 1]
virginica_samples = dataset_df[dataset_df['species'] == 2]

# Undersample each class to have the same number of samples
setosa_undersampled = setosa_samples.sample(target_samples, random_state=42)
versicolor_undersampled = versicolor_samples.sample(target_samples, random_state=42)
virginica_undersampled = virginica_samples.sample(target_samples, random_state=42)

undersampled_df = pd.concat([setosa_undersampled, versicolor_undersampled, virginica_undersampled])


undersampled_df = undersampled_df.sample(frac=1, random_state=42).reset_index(drop=True)

print("Undersampled class distribution:")
print(undersampled_df['species'].value_counts())
```

```python
undersampled_df.to_csv('undersampled_df.csv', index=False)
```
